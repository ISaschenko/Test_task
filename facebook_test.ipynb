```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('dataset_Facebook.csv', sep=";")
```

#### Для начала я находил  значения по столбцам для всего массива.

Для данных в столбоцах Type, Category, Post Month, Post Weekday, Post Hour, Paid я не стал рассчитывать значения, так как они не имеют ценности.

Post Month, Post Weekday, Post Hour - это временные признаки события(фотографии, видео, статуса или ссылки) и, разве что мода может дать нам каку-то информацию, например в какой день недели или месяц чаще публикуются важные события.


Type, Category - категориальные признаки.

Paid - бинарный признак(опять же не имеет смысла рассчитывать среднее, макс/мин, медиану).

```python
notValid = ["Type","Category","Post Month","Post Weekday", "Post Hour","Paid"]
```

```python
for i in data.columns:
    if i not in notValid:
        print(i.upper() ,'\n','Среднее значение:',data[i].mean(), "Максимальное:", data[i].max(),"Минимальное:", data[i].min(),
            "Медиана:",data[i].median(),"\nМода:", data[i].mode(), "\n")
```

#### Теперь я найду те же значения отдельно по типам:

```python
types = set()
for i in data['Type']:
    if i not in types:
        types.add(i)
types
```

```python
for t in types:
    tempdata = data[data['Type'] == t]
    print("Значения для '",t.upper(),"':")
    for i in data.columns:
        if i not in notValid:
            print(i.upper() ,'\n','Среднее значение:',tempdata[i].mean(), "Максимальное:", tempdata[i].max(),
                  "Минимальное:", tempdata[i].min(), "Медиана:",tempdata[i].median(),"\nМода:", tempdata[i].mode(), "\n")
    
        
```

### Самый популярный обьект в выборке.

Для определения самого популярного обьекта в выборке нужно посчитать коэфициент популярности, который состоит из суммы некатегориальных признаков с определенными весами. Вопрос в том, какие веса подобрать для признаков. Попробуем дать всем признакам одинаковые веса.

```python
import operator
data1 = data.drop(notValid, axis = 1)
data1.fillna(0, inplace = True)
weight = 1/len(data1.columns)
d = {}
for i in range(len(data)):
    pop = 0
    for c in data1.columns:
            pop += weight*int(data1[c][i])
    d[i] = pop
max(d.items(), key=operator.itemgetter(1))[0] 
```
